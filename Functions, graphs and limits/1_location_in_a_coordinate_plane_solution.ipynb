# Location in a Coordinate Plane Solution

## Activity 1

Plot the points $(2,3), (0,6), (-5,1), (4,-3), (0.5,-4), (-1,-2), (-1.5,0), (6,-5)$.

```python
from sympy import *
from sympy.geometry import *

# Define the symbols
x, y = symbols('x y')

# Define the point A
A = Point(2, 3)

# Plot the point A as intersections of the lines x = 2 and y = 3
px = plot_implicit(Eq(x, A.x), (x,-10,10), (y,-10,10), show=False)
py = plot_implicit(Eq(y, A.y), (x,-10,10), (y,-10,10), show=False)
px.extend(py)
px.show()
```

```python
# Define the points
A = Point(2, 3)
B = Point(0, 6)
C = Point(-5, 1)
D = Point(4, -3)
E = Point(0.5, -4)
F = Point(-1, 2)
G = Point(-1.5, 0)
H = Point(6, -5)

# Define the list of points
points = [A, B, C, D, E, F, G, H]

# Plot the points
for point in points:
    print("Plotting the point: ", point)
    px = plot_implicit(Eq(x, point.x), (x,-10,10), (y,-10,10), show=False)
    py = plot_implicit(Eq(y, point.y), (x,-10,10), (y,-10,10), show=False)
    px.extend(py)
    px.show()
```

## Activity 3

Find the midpoint of the line connecting the two points:

- $(5,6), (9,2)$
- $(0,0), (-4,8)$
- $(-10,4), (-6,8)$
- $(7,-9), (-3,5)$

```python
# Define the points
A, B = Point(5, 6), Point(9, 2)
C, D = Point(0, 0), Point(-4, 8)
E, F = Point(-10, 4), Point(-6, 8)
G, H = Point(7, -9), Point(-3, -5)

# Define the list of couples of points
points = [(A, B), (C, D), (E, F), (G, H)]

for couple in points:
    # Define the midpoint of the points
    midpoint = Segment(couple[0], couple[1]).midpoint
    print(midpoint)
```

## Activity 4

Translate the triangle whose vectices are $(1,3)$, $(2,4)$ and $(5,6)$ three units to the right and four units up.

Find the coordinates of the translated vertices.

```python
# Define the points
A = Point(1, 3)
B = Point(2, 4)
C = Point(5, 6)

# Define the triangle
triangle = Triangle(A, B, C)
triangle.vertices
```

```python
# Translate the triangle
translated_triangle = triangle.translate(3, 4)
translated_triangle.vertices
```

## Activity 5

Translate the rectangle whose vertices are $(-2,1)$, $(-1,2)$, $(1,0)$ and $(0,-1)$ four units to the right and one unit down.

Find the coordinates of the translated vertices.

```python
# Define the points
A = Point(-2, 1)
B = Point(-1, 2)
C = Point(1, 0)
D = Point(0, -1)

# Define the rectangle
rectangle = Polygon(A, B, C, D)
rectangle.vertices
```

```python
# Translate the rectangle
translated_rectangle = rectangle.translate(4, -1)
translated_rectangle.vertices
```
